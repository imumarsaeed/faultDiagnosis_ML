```python
import os
import pandas as pd
```

```python
df = pd.read_excel(r"C:\Users\MCSL-user\Pictures\1.0_constant_FaultDiagnosis_multihop_indoor_dataset.xlsx",
                  sheet_name='normal')
df.head()
```

```python
drift_value = 1                   # s = sig_n = nb0 --- b0=constt

df_faulty1 = df[['1_T1','3_H1']]
drift1 = df_faulty1.head(300)
cell_n1 = drift1.add(drift_value * 1)

df_faulty2 = df[['5_T1','7_H1']]
drift2 = df_faulty2.head(300)
cell_n2 = drift2.add(drift_value * 2)

df_faulty3 = df[['9_T1','11_H1']]
drift3 = df_faulty3.head(300)
cell_n3 = drift3.add(drift_value * 3)

df_faulty4 = df[['13_T1','15_H1']]
drift4 = df_faulty4.head(300)
cell_n4 = drift4.add(drift_value * 4)


df_normal = df[['2_T2','4_H2','6_T2','8_H2','10_T2','12_H2','14_T2','16_H2']]
finalnormal = df_normal.head(300)

finaldf = pd.concat([cell_n1, cell_n2, cell_n3, cell_n4, finalnormal], axis=1)
finaldf1 = finaldf[['1_T1','2_T2','3_H1','4_H2','5_T1','6_T2','7_H1','8_H2','9_T1','10_T2','11_H1','12_H2','13_T1','14_T2','15_H1','16_H2']]
```

```python
drift_value = 1                  # s = sig_n = nb0 --- b0=constt

df_faulty1 = df[['2_T2','4_H2']]
drifT2 = df_faulty1.tail(300)
cell_n1 = drifT2.add(drift_value * 1)

df_faulty2 = df[['6_T2','8_H2']]
drift2 = df_faulty2.tail(300)
cell_n2 = drift2.add(drift_value * 2)

df_faulty3 = df[['10_T2','12_H2']]
drift3 = df_faulty3.tail(300)
cell_n3 = drift3.add(drift_value * 3)

df_faulty4 = df[['14_T2','16_H2']]
drift4 = df_faulty4.tail(300)
cell_n4 = drift4.add(drift_value * 4)


df_normal = df[['1_T1','3_H1','5_T1','7_H1','9_T1','11_H1','13_T1','15_H1']]
finalnormal = df_normal.tail(300)

finaldf = pd.concat([cell_n1, cell_n2, cell_n3, cell_n4, finalnormal], axis=1)
finaldf2 = finaldf[['1_T1','2_T2','3_H1','4_H2','5_T1','6_T2','7_H1','8_H2','9_T1','10_T2','11_H1','12_H2','13_T1','14_T2','15_H1','16_H2']]
```

```python
extreme_final = pd.concat([finaldf1, finaldf2])
extreme_final
```

```python
extreme_final.to_excel(r'C:\Users\MCSL-user\Pictures\drift.xlsx', sheet_name='drift')
```
